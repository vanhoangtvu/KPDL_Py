```python
# Ensure this cell is run first to define dog_images and cat_images
import cv2
import os

def load_images_from_folder(folder):
    data = []
    for filename in os.listdir(folder):
        if filename.lower().endswith('.jpg'):
            img = cv2.imread(os.path.join(folder, filename))
            data.append(img)
    return data

dog_images = load_images_from_folder('training/dog')
cat_images = load_images_from_folder('training/cat')
```

```python
# This cell should be run after the cell above
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random # Import the random module

plt.figure(figsize=(2, 2))
# Generate a random index within the range of dog_images length
random_dog_index = random.randint(0, len(dog_images) - 1)
plt.imshow(cv2.cvtColor(dog_images[random_dog_index], cv2.COLOR_BGR2RGB))

plt.figure(figsize=(2, 2))
# Generate a random index within the range of cat_images length
random_cat_index = random.randint(0, len(cat_images) - 1)
plt.imshow(cv2.cvtColor(cat_images[random_cat_index], cv2.COLOR_BGR2RGB))
```

```python
# use it as a magic tool
from transformers import CLIPProcessor, CLIPModel
import torch

clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")

# CLIP
def magic_function(images):
    inputs = clip_processor(images=images, return_tensors="pt", padding=True)
    with torch.no_grad():
        features = clip_model.get_image_features(**inputs)
    return features.numpy()

dog_images_features = magic_function(list(dog_images))
cat_images_features = magic_function(list(cat_images))
```
